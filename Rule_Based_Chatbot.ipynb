```python
class RuleBot:
    # Potential Negative Responses
    negative_responses = ("no", "nope", "nah", "not a chance", "sorry")
    # Exit conversation keywords
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later", "I'll talk to you later")
    general_commands = ("Nice","hmm","I see", "Okk","alright","great")
    # Random starter questions
    random_questions = (
        "Why are you here?",
        "Are there many humans like you?",
        "What do you consume for sustenance?",
        "Does Earth have a leader?",
        "Is there any intelligent life on this planet?",
        "What planets have you visited?",
        "What technology do you have on this planet?",
        "How do you explore space?",
        "What kind of art do you create?",
        "What is your planet's climate like?",
        "What else can you tell me about Earth’s history?",
        "How does Earth's atmosphere support life?",
        "What other amazing plants or animals are there on Earth?",
        "Can you tell me more about the ecosystems where these plants and animals live?",
    )

    def __init__(self):
        self.alienbabble = {
            'describe_planet_intent': r'.*\s*your planet.*',
            'answer_why_intent': r'.*\s* why',
            'about_earth_technology_intent': r'.*\s*technology',
             'about_earth_technology_intent': r'.*\s*technologies',
            'about_earth_protection': r'.*\s*protect',
            'Earth_facts_Intent': r'.*\s*facts',
            'Flora_Fauna_Intent': r'.*\s*plants and animals',
            'Flora_Fauna_Intent': r'.*\s*animals',
            'Flora_Fauna_Intent': r'.*\s*plants'
        }

    def greet(self):
        self.name = input("Hi there! What’s your name? I’d love to get to know you better:)\n")
        will_help = input(f"Hi {self.name}, I am a friendly bot. I'm curious to learn about your planet Earth. Will you help me learn about your planet?\n")
        if will_help.lower() in self.negative_responses:
            print("OK, have a nice Earth day!")
            return
        self.chat()

    def make_exit(self, reply):
        for command in self.exit_commands:
            if reply.lower() == command:
                print("Okay, have a nice Earth day!")
                return True
        return False

    def continue_conv(self, reply):
        return reply.lower() in self.general_commands

    def chat(self):
        while True:
            reply = input("\nAsk me anything about earth: about your planet, facts about the earth, about plants and animals,technology used, ways to protect earth or say 'exit' to end our chat:\n"
            )
            if self.make_exit(reply):
                break
            response = self.match_reply(reply)
            print(response)

    def match_reply(self, reply):
    # Iterate over the dictionary of intents and patterns
     for key, pattern in self.alienbabble.items():
        # Use re.match to check if the pattern matches the reply
        if re.match(pattern, reply, re.IGNORECASE):
            # If a match is found, call the corresponding method dynamically using getattr
            return getattr(self, key)()
    # If no match is found, return a default response
     return self.no_match_intent()

    def describe_planet_intent(self):
        responses = (
            "Sure!Earth, the blue jewel of our solar system, is home to diverse life forms and dynamic ecosystems.\n I hope this answers your question",
            "Here you go!Earth's diverse ecosystems, from lush rainforests to arid deserts, showcase the remarkable variety of life and landscapes that thrive on our unique planet.\nI hope you found this helpful",
            "Right away!Earth is home to many beautiful plants. Trees, flowers, and bushes grow all around us, giving us fresh air to breathe and lovely colors to see.\nI hope this answers your question",
            "Here you go!Our planet has lots of amazing animals. From tiny ants to big elephants, each animal plays a special part in our world, making it full of life and wonder.\n Let me know if you need more information",
            "Sure!Nature on Earth is like a giant playground. We have tall mountains, deep oceans, and big forests. Everywhere you look, there is something wonderful to see.\nFeel free to ask if you have any more questions",
            "Right away!People live on Earth too, just like us. We build homes, go to school, play games, and care for each other. Together, we make Earth a very special place.\n I hope this helps!",
            "Absolutely!Plants are like Earth's green hair. They grow everywhere, from tiny flowers to big trees, and they help us breathe by making fresh air.\nI hope you find this informative."
        )
        return random.choice(responses)

    def answer_why_intent(self):
        responses = (
            "Absolutely!I come in peace to explore Earth.\n",
            "Right away,I am here to collect data on your planet and its inhabitants.\n",
            "Here you go!I heard the coffee is really good.\n",
            "Absolutely!I want to learn about Earth because it’s full of life and mystery!\n",
            "Well,the idea of meeting new friends on Earth makes me very curious and happy!\n",
            "Humans are fascinating beings, and I want to understand how they live and think.\n",
            "I heard Earth has delicious food, and I want to know all about it!\n",
            "Your planet has so much history and culture; it's like a giant storybook I want to read.\n"
        )
        return random.choice(responses)

    def about_earth_technology_intent(self):
        responses = (
            "Right away, On Earth, we use computers to help us work, play, and connect with friends and family far away.\nI hope you found this helpful",
            "Absolutely!Smartphones are like tiny computers we carry in our pockets, letting us talk to anyone and find information quickly.\nLet me know if you need more information.",
            "Sure, here you go.The Internet is a magical web that connects all our computers, letting us share and learn new things from all over the world.\nFeel free to ask if you have any more questions",
            "Absolutely!Robots are machines that can do tasks for us, like cleaning our homes or helping in factories.\n I hope this helps!",
            "You got it!Solar panels capture sunlight and turn it into electricity, giving us a clean way to power our homes.\nI hope this answers your question",
            "Space rockets take astronauts on exciting trips to explore space, helping us learn more about the universe.\nI hope this answers your question",
            "Smart home devices, like voice assistants, can control lights and play music with just a few words, making homes more convenient.\nI hope this helps",
            "Absolutely!Space telescopes help scientists look far into the universe and learn more about distant stars and planets.\nI hope this helps"
        )
        return random.choice(responses)

    def about_earth_protection(self):
        responses = (
            "Here you go!By reducing waste and reusing items, we help keep our planet clean and reduce pollution.\nI hope this helps!",
            "Absolutely!Conserving Earth means taking small steps every day, like picking up litter and using less water, to keep our planet beautiful and healthy.\nLet me know if you need more information",
            "Absolutely!Planting trees and taking care of nature helps keep our air clean and provides homes for animals.\nI hope this helps!",
            "Right away!When we take care of Earth, we keep our air, water, and land clean and beautiful for everyone to enjoy.\nI trust this is the information you were looking for",
            "Right away!The 5 R's help us protect Earth: Reduce what we use, Reuse what we can, Recycle materials, Repair things instead of throwing them away, and Respect nature by not wasting resources.\nI hope this helps!",
            "Sure, here you go!By practicing these 5 R's, we help keep Earth healthy and beautiful, ensuring it remains a wonderful place to live for all living things.\n",
            "Most definitely!Conserving Earth is like giving our planet a big hug, making sure it stays healthy and happy for all living things.\nIf you need more details, just let me know",
            "Absolutely!When we all do our part to protect the planet, we help keep it beautiful and safe for everyone to live and thrive.\nFeel free to ask if you have any more questions",
            "Absolutely!Taking small steps like turning off lights, using less water, and picking up litter can make a big difference in protecting our Earth.\nFeel free to ask if you have any more questions",
        )
        return random.choice(responses)

    def Earth_facts_Intent(self):
        responses = (
            "Of course!Earth formed 4.6 billion years ago from gas and dust around the Sun.\n I hope this helps!",
            "Certainly!Around 700 million years ago, Earth experienced extreme glaciations, covering the entire planet in ice i.e., 'Snowball Earth'.\nLet me know if you need more information",
            "Absolutely!Oceans cover about 71% of Earth’s surface.\nLet me know if you need more information",
            "Here you go!Around 541 million years ago, life on Earth experienced a creative burst—the Cambrian Explosion.\nI hope you find this informative.",
            "Most definitely!Earth’s crust is divided into tectonic plates that move around. Sometimes they collide, forming supercontinents.\n",
            "Here you go!Earth is approximately 4.6 billion years old.\nLet me know if you need more information",
            "Absolutely!Scientists estimate that Earth will remain habitable for another 1-2 billion years\n I hope this helps!",
            "Absolutely!Our planet is home to five major oceans: Pacific, Atlantic, Indian, Southern, and Arctic\nI hope you find this informative",
            "Most definitely!Earth’s rotation gradually slows due to the Moon’s pull, due to which we experience longer days billion years later :)\nLet me know if you need more information"
        )
        return random.choice(responses)

    def Flora_Fauna_Intent(self):
        responses = (
            "Most definitely!Did you know that the Amazon Rainforest is often called the lungs of the Earth? It's home to around 390 billion individual trees across 16,000 species!\nI hope this helps!",
            "Sure, here you go!One fascinating creature is the axolotl, a type of salamander that can regenerate its limbs, spinal cord, heart, and other organs throughout its life.\nLet me know if you need more information.",
            "Absolutely!The Rafflesia arnoldii, also known as the corpse flower, is the largest flower on Earth. It can grow up to 3 feet in diameter and smells like rotting flesh!\n",
            "Absolutely!The blue whale is the largest animal ever known to have lived on Earth, reaching lengths of up to 100 feet and weights of around 200 tons!\nLet me know if you need more information.",
            "Right away!The Venus flytrap is a carnivorous plant that snaps shut on its prey, primarily insects, with lightning speed—within less than a second!\nFeel free to ask if you have any more questions",
            "Right away!There are more species of beetles on Earth than any other type of animal. Scientists estimate there are about 400,000 different species of beetles!\nFeel free to ask if you have any more questions",
            "Most definitely!Did you know that there are more than 400,000 species of flowering plants on Earth? They come in an incredible variety of shapes, sizes, and colors, contributing to the planet's biodiversity.\nI hope this helps!",
            "Certainly!Earth's coral reefs are often referred to as 'rainforest of the sea' because of their incredible biodiversity. They support over 4,000 species of fish and countless other marine organisms.\nI hope this helps!"
        )

        return random.choice(responses)

    def no_match_intent(self):
        responses = (
            "I'm not sure what you mean. Could you ask something else?",
            "I don't understand. Can you try asking differently?\n",
            "I'm not sure how to answer that. Can you ask another way?\n",
            "I see. Can you elaborate?\n",
            "I didn't get that. Can you rephrase\n",
            "That's interesting, but I don't have a response for that. Can you ask something else?\n"
        )
        return random.choice(responses)

# Create an instance of RuleBot and start the interaction
AlienBot = RuleBot()
AlienBot.greet()
```
